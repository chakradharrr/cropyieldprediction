```python
import pandas as pd 
d=pd.read_excel("final catgry.xlsx")
d
```

```python
d=d.drop(['Unnamed: 0.1', 'Unnamed: 0'],axis=1)
d
```

```python
d.columns
```

```python
x=d[['Rainfall(mm)', 'Minimum Temparature(C)','Maximum Temperature (C)','Irrigation(MBGL)','Season_Kharif', 'Season_rabi','Season_Kharif', 'Season_rabi', 'Crop_Groundnut', 'Crop_Maize','Crop_Moong(Green Gram)', 'Crop_Rice', 'Crop_cotton(lint)']]
y=d['Yield (Bales/Hectare)']
x,y
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
rf.fit(x,y)
```

```python
y_pred=rf.predict(x)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mean_squared_error(y,y_pred)
```

```python
r2_score(y,y_pred)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
rf=RandomForestRegressor()
rf.fit(x_train,y_train)
```

```python
y_pred=rf.predict(x_test)
print("r2_score",r2_score(y_test,y_pred))
print("mean_sq_err",mean_squared_error(y_test,y_pred))
```

```python
from sklearn.ensemble import BaggingRegressor
bg=BaggingRegressor()
bg.fit(x_train,y_train)
```

```python
y_pred=bg.predict(x_test)
print("r2_score",r2_score(y_test,y_pred))
print("mean_sq_err",mean_squared_error(y_test,y_pred))
```
