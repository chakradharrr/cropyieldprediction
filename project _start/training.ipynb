```python
import numpy as np

for i in df.columns:
    if(df[i].dtype!="object"):
        q1=np.percentile(df[i],25)
        q3=np.percentile(df[i],75)
        iqr=q3-q1
        ll=q1-1.5*iqr
        ul=q3+1.5*iqr
        print(df.loc[(df[i]<ll) | (df[i]>ul),i])
```
