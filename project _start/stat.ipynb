```python
import pandas as pd
import numpy as np
from sklearn.ensemble import  RandomForestRegressor
df=pd.read_excel("final1.xlsx")
df
```

```python
df.dropna(subset=['Yield (Bales/Hectare)'],inplace=True)
df.isnull().sum()
```

```python
df.head(10)
a=df['Year']
df.dropna(subset=['Year'],inplace=True)
df
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor

# load data


# replace 'Season' column with numeric codes
df['Season'] = df['Season'].replace({'Kharif': 0, 'Rabi': 1})

# split data into training and testing sets
train = df[df['Minimum Temperature'].notna()] # rows with non-missing values
test = df[df['Minimum Temperature'].isna()] # rows with missing values

# define input and output columns
input_cols = ['Year', 'Season', 'District']
output_cols = ['Minimum Temparature(C)', 'Maximum Temperature(c)', 'Irrigation']

# train a random forest model for each output column
for col in output_cols:
    # split training data into input and output
    X_train = train[input_cols]
    y_train = train[col]

    # train random forest model
    model = RandomForestRegressor(n_estimators=100, random_state=42)
    model.fit(X_train, y_train)

    # use model to fill in missing values in test data
    X_test = test[input_cols]
    y_test = model.predict(X_test)
    test[col] = y_test

# combine training and testing sets back into one dataframe
df_filled = pd.concat([train, test])

# save filled data to file

```
