```python
import numpy as np
l=['Minimum Temparature(C)','Maximum Temperature (C)']
for i in l:
    if(d[i].dtype!="object"):
        q1=np.percentile(d[i],25)
        q3=np.percentile(d[i],75)
        iqr=q3-q1
        ll=q1-1.5*iqr
        ul=q3+1.5*iqr
        print(d.loc[(d[i]<ll) | (d[i]>ul),i])
```
