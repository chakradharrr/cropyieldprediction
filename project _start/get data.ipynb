```python
import datetime
current_date = datetime.date.today()
if current_date.month >= 6 and current_date.month <= 10:
    Season = 'Kharif'
else:
    Season = 'rabi'

district=input("Select district")
data = d[(d['District'] == district) & (d['Season'] == Season)]
data
```

```python
avg_rainfall=data['Rainfall(mm)'].mean()
avg_min_temp=data['Minimum Temparature(C)'].mean()
avg_max_temp=data['Maximum Temperature (C)'].mean()
print("avg_rainfall:",avg_rainfall)
print("avg_min_temp:",avg_min_temp)
print("avg_max_temp:",avg_max_temp)
```
